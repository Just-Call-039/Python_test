```python
import pandas as pd

df_1 = pd.read_csv("last 3 steps.csv", sep=";")
df_2 = pd.read_csv("last 2 steps.csv", sep=";")
df_3 = pd.read_csv("Total_data.csv", sep=";")

column_names = ["steps", "positive", "negative", "chance", "count_3_steps", "count_2_steps"]

df_4 = df_3.merge(df_1, how='left', left_on="steps", right_on="last_steps", left_index=False, right_index=False)
del df_4["last_steps"]

df_5 = df_4.merge(df_2, how='left', left_on="steps", right_on="last_steps", left_index=False, right_index=False)
del df_5["last_steps"]

df_5.to_csv("Merge_1.csv", index=False, na_rep='Unknown', sep=";", header=column_names)

df_5.head(30)
```
