```python
import pandas as pd
import time

start_time = time.time()

df = pd.read_csv("group.csv", sep=",")

bad_list = ["2", "111", "1", "261", "262", "250", "3", "4", "5",
             "6", "7", "8", "9", "10", "11", "102", "103", "104",
             "105", "106", "107", "108", "109", "113"]
bad_list = [int(i) for i in bad_list]

# print(bad_list)

# print(df.head())

#       .   .
k = 0
my_dict = {}

for now in df.values:
    #  ,      0.
    list_steps = now[0].strip("[]").split(",")
    #  ,        1.
    count = now[1]
    # print(list_steps)
    #     ,      3.
    if len(list_steps) >= 3:
        list_steps = [int(i.strip("''")) for i in list_steps]
        #      .    ,    .
        #    ,    - .
        # first -   .
        first = tuple(list_steps[-3:-1])
        # second -  ,     .
        second = list_steps[-1]
        # print(f"list_steps {list_steps} first {first} second {second}")
        #      .
        #   -   .
        #  - . 
        #    0 -   .
        #    1 -   .
        # [+, -]: [good, bad]
        if first in my_dict:
            if second in bad_list:
                my_dict[first][1] += count
            else:
                my_dict[first][0] += count
        else:
            if second in bad_list:
                my_dict[first] = [0, count]
            else:
                my_dict[first] = [count, 0]
    #   1'000'000 .
    k += 1    
    if k == 10000000:
        break

# print(my_dict)

#   ,        (  ).
result_dict = dict()

for now in my_dict:
    #    ,    .
    chance = round(my_dict[now][1] / (my_dict[now][0] + my_dict[now][1]), 2)
    #  .  -     ,   .
    #    -  ,    .
    result_dict[now] = chance

print(result_dict)
print()
print(f"  : {len(result_dict)} .")

#  1'000'000      19.42 seconds.
print(f"  {k}  : {round(time.time() - start_time, 2)} .")
```
