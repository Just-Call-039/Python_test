```python
# Программа обрабатывает файл с заголовком: список шагов, количество таких шагов. Разделитель - ";".
# Записывает результирующий файл: 
# список шагов, количество хороших шагов, количество плохих шагов, итоговое количество, коэффициент. Разделитель - ";".

import pandas as pd
import time


# Рассчет коэффициента.
def ratio(good, bad):
    # ratio = bad / (good + bad)
    r = bad / (good + bad)
    return round(r, 2)


good_step = 990

k = 0
my_dict = {}
my_list = []
open_f = input("Название файла и расширение на чтение: ")
my_file = input("Название файла и расширение на запись (например: data_1.csv, data_2.csv, data_3.csv и т.д.): ")

# Для подсчета выполнения программы.
start_time = time.time()

df = pd.read_csv(open_f, sep=";")
# print(df.value_counts())

# Открытие файла на запись.
with open(my_file, "w") as data:
    # Итерация по значениям.
    for now in df.values:
        k += 1
        # Список шагов.
        list_steps = now[0].strip("[]").split(",")
        # Количество таких шагов.
        count = now[1]
        
        # Берем строки только >= 6.
        if len(list_steps) >= 6:
            list_steps = [int(i.strip("''")) for i in list_steps]
            start = -6
            end = -1
            # print(list_steps, count)
            
            # Итерация по строке и отбор шагов для списка. Длина каждого списка - 5.
            while abs(start) <= len(list_steps):
                first = tuple(list_steps[start:end])
                # Смещение на 1 влево.
                second = list_steps[end]
                start -= 1
                end -= 1
                # Добавление значений в словарь.
                # last_steps: [good, bad, total, ratio]
                if first in my_dict:
                    # На 0 позиции - хороший шаг.
                    if second == good_step:
                        my_dict[first][0] += count
                    # На 1 позиции - плохой шаг.
                    else:
                        my_dict[first][1] += count
                # Создание ключа и значения.
                else:
                    # На 0 позиции - хороший шаг. И заглушка для total, ratio.
                    if second == good_step:
                        my_dict[first] = [count, 0, 0, 0]
                    # На 1 позиции - плохой шаг. И заглушка для total, ratio.
                    else:
                        my_dict[first] = [0, count, 0, 0]

        # if k == 100:
        #    break
    
    # Запись заголовка в файл.
    data.write("last_steps; good; bad; total; ratio\n")
    
    # Итерация по словарю.
    for now in my_dict:
        # Итоговое число шагов - сумма из положительных и отрицательных.
        total = my_dict[now][0] + my_dict[now][1]
        # Коэффициент.
        r = ratio(my_dict[now][0], my_dict[now][1])
        # Запись в словарь итогового числа шагов и коэффициента.
        my_dict[now][2], my_dict[now][3] = total, r
        # Создание списка из ключей, у которых итоговое число шагов меньше 100.
        if total < 100:
            my_list.append(now)
    
    # Итерация по списку и удаление такой пары (ключ: значение) из словаря.
    for now in my_list:
        del my_dict[now]
    
    # Итерация по словарю и запись в файл.
    for now in my_dict:
        data.write(f"{now}; {my_dict[now][0]}; {my_dict[now][1]}; {my_dict[now][2]}; {my_dict[now][3]}\n")

# print(my_dict)
print(f"Длина словаря составляет: {len(my_dict)} значений.")
print(f"Время обработки {k} строк составило: {round(time.time() - start_time, 2)} сек.")
```
